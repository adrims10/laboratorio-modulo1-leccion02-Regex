Dado el archivo `datos_artistas.txt` usa las **expresiones regulares** para extraer los siguientes datos, con el código que encontrarás en la siguiente celda:

```python
with open('Datos/datos_artistas.txt', 'r',encoding = "utf-8") as archivo:
    contenido = archivo.read()
    listadecanciones= contenido.split('./')
    # Aquí puedes trabajar con el contenido del archivo

print(contenido[:10000])
```

#### 1.1 La información de cada artista(nombre, correo de contacto, número de contacto y sitio web).

```python

listatotal=[]

for cancion in listadecanciones:
    patronnombre = r'\w+'
    patroncorreo= r'\s(.+\.\w{2,3})\s-'
    patronnumero= r'(-\d{1,3}\s?\d+)\s'
    patronw=r'\s(https?.+\.\w{2,3})\s'

    artistanombre = re.match(patronnombre,cancion).group()
    correo = re.findall(patroncorreo,cancion)[0]
    numero = re.findall(patronnumero,cancion)[0]
    web = re.findall(patronw,cancion)[0]
    listatotal.append([artistanombre,correo,numero,web])

listatotal

```

#### 1.2 La información de cada canción(título, duración y fecha de lanzamiento).

```python

canciones=[]

for cancio in listadecanciones:
    
    patron_titulo = r'\.\w{2,3}\s([\w+\s]+)'
    patron_duracion = r'-\s(\d{1,2}:\d+)'
    patron_lanzamiento = r'-\s\d{1,2}:\d+[\s,]?(\d{1,4})\s'

    titulo = re.findall(patron_titulo,cancio)[0]
    duracion = re.findall(patron_duracion,cancio)[0]
    lanzamiento = re.findall(patron_lanzamiento,cancio)[0]
    
    
    canciones.append([titulo,duracion,lanzamiento])

canciones
```

#### 1.3 Almacena la información en una lista de diccionarios donde cada diccionario contenga los datos de un artista y su canción asociada.

```python

Listadedatos = []

for contadorvueltas ,artista in enumerate(listatotal):
        Listadedatos.append({
                'artista': artista,
                'cancion':canciones[contadorvueltas]   
        })

Listadedatos
    
```
